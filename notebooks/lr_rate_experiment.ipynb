```python
import plotly.express as px
import torch
import torch.nn as nn
from torch.optim.lr_scheduler import OneCycleLR
```

```python
model = nn.Linear(20, 10)

sample_input = torch.rand(20)
print(sample_input)
model_output = model(sample_input)
steps = 9120
lr = 2e-5
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=lr)
scheduler = OneCycleLR(optimizer, max_lr=7e-5, steps_per_epoch=304, pct_start=0.15, epochs=30,
                    anneal_strategy='cos',div_factor=100)
```

```python
lrs = []
for i in range(steps):
    optimizer.step()
    lrs.append(optimizer.param_groups[0]["lr"])
    #     print("Factor = ",i," , Learning Rate = ",optimizer.param_groups[0]["lr"])
    scheduler.step()
```

```python
px.line(lrs)
```

```python
epoch = 0
for index, val in enumerate(lrs):
    if index % 304 == 0:
        print("epoch:", epoch)
        epoch+=1
    print(index, val)
```
